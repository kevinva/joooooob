```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 边缘检测
image = cv2.imread('./WechatIMG263.jpeg')
imageBlur = cv2.GaussianBlur(image, (3, 3), 1)
imageGray = cv2.cvtColor(imageBlur, cv2.COLOR_RGB2GRAY)
edges = cv2.Canny(imageGray, 60, 100)

imageOut = cv2.cvtColor(edges, cv2.COLOR_GRAY2RGB)
# cv2.imshow('test', imageOut)
plt.imshow(imageOut)
```

```python
# 梯度图： sobel算子
# img_gradient_X_8U = cv2.Sobel(image, -1, 0, 1)  # y方向的一阶导数
img_gradient_X_8U = cv2.Sobel(image, -1, 1, 0)  # x方向的一阶导数
img_gradient_X_64F = cv2.Sobel(image, cv2.CV_64F, 1, 0)

img_gradient_X_64Fto8U = cv2.convertScaleAbs(img_gradient_X_64F)

# cv2.imshow('X_gradient_8U', img_gradient_X_8U)
# cv2.waitKey()
# cv2.destroyAllWindows()
plt.figure(figsize=(16, 9))
plt.imshow(img_gradient_X_8U)
```

```python
# 梯度图像计算

grad_x = cv2.Sobel(image, -1, 1, 0)
grad_y = cv2.Sobel(image, -1, 0, 1)

grad = cv2.addWeighted(grad_x, 0.5, grad_y, 0.5, 0)
plt.figure(figsize=(16, 9))
plt.imshow(grad)
```
